# Notebook 4.1 -- Composing networks (Julia Flux Version)

The purpose of this notebook is to understand what happens when we feed one neural network into another using the Flux.jl library.

```julia
using Flux
using Plots
```

### Define the Shallow Network
In Flux, a shallow network with 3 hidden units is represented as a `Chain` of two `Dense` layers.

```julia
function shallow_1_1_3(W1, b1, W2, b2)
    return Chain(
        Dense(W1, b1, relu), # Hidden layer
        Dense(W2, b2)        # Output layer
    )
end
```

### Define Parameters
We use the exact parameters provided in the original Python source.

```julia
# Network 1 parameters
n1_W1 = reshape([-1.0, 1.0, 1.0], 3, 1)
n1_b1 = [0.0, 0.0, -0.67]
n1_W2 = reshape([-2.0, -3.0, 9.3], 1, 3)
n1_b2 = [1.0]

# Network 2 parameters
n2_W1 = reshape([-1.0, 1.0, 1.0], 3, 1)
n2_b1 = [-0.6, 0.2, -0.5]
n2_W2 = reshape([-1.0, -1.5, 2.0], 1, 3)
n2_b2 = [0.5]

net1 = shallow_1_1_3(n1_W1, n1_b1, n1_W2, n1_b2)
net2 = shallow_1_1_3(n2_W1, n2_b1, n2_W2, n2_b2)

x = reshape(collect(-1.0:0.001:1.0), 1, :) # Input range [-1, 1]
```

### Run Composition
We feed the output of the first network as input to the second.

```julia
net1_out = net1(x)
net2_out = net2(x)

# Composed output: Net2(Net1(x))
net12_out = net2(net1_out)

# Visualize
p1 = plot(vec(x), vec(net1_out), color=:red, title="Net 1", xlims=(-1,1), ylims=(-1,1))
p2 = plot(vec(x), vec(net2_out), color=:blue, title="Net 2", xlims=(-1,1), ylims=(-1,1))
p3 = plot(vec(x), vec(net12_out), color=:green, title="Composition", xlims=(-1,1), ylims=(-1,1))

plot(p1, p2, p3, layout=(1,3), size=(1000, 300))
```

### Deep Composition (3 Layers)
Passing the result into multiple copies to see how linear regions multiply.

```julia
# net123_out = net2(net1(net1(x)))
net123_out = net2(net1(net1_out))
plot(vec(x), vec(net123_out), color=:black, title="Three Layer Composition", lw=2)
```
